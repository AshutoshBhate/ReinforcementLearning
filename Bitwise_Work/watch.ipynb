```python
import gymnasium as gym
from stable_baselines3 import PPO
import torch
import numpy as np
from environment import PbRLWrapper
from reward_model import RewardNetwork
from buffers import TrajectoryBuffer

def watch():
    print("Loading models for demonstration")

    # Recreate the environment with Render Mode turned on
    raw_env = gym.make("CartPole-v1", render_mode="human")
    
    dummy_reward_net = RewardNetwork(4) 
    dummy_buffer = TrajectoryBuffer()
    env = PbRLWrapper(raw_env, dummy_reward_net, dummy_buffer)

    # Load the trained Agent
    try:
        model = PPO.load("ppo_pbrl_agent", env=env)
        print("Agent loaded successfully!")
    except:
        print("Error: Could not find 'ppo_pbrl_agent.zip'. Did you run main.py?")
        return

    # Play the game
    obs, info = env.reset()
    for _ in range(1000):
        
        action, _states = model.predict(obs, deterministic=True)
        
        obs, reward, terminated, truncated, info = env.step(action)
        
        if terminated or truncated:
            obs, info = env.reset()

    env.close()

if __name__ == "__main__":
    watch()
```
